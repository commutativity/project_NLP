## Formeln

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def my_tokenizer(txt):
    doc = nlp(txt)
    tokens = []
    for token in doc:
        if token.pos_ == 'PUNCT' or token.pos_ == 'SPACE':
                continue
        tokens.append(token.text)    
    #s = re.findall("[A-Za-zßäöüÄÖÜ]+",txt)
    #for i in range(len(s)):
    #    s[i] = s[i].upper().replace("Ä","AE").replace("Ö","OE").replace("Ü","UE")
    return ' '.join(tokens)
```

## Metadaten

```python
corpus = [item['description'] for item in desc]
```

```python
docs = []
length_desc = []
```

```python
for item in corpus:
    txt = my_tokenizer(item)
    docs.append(txt)
    length_desc.append(len(txt.split(' ')))
```

```python
n_descriptions = len(corpus)
```

```python
max_id = 0
min_id = 0
max_value = 0
min_value = 1000
sum_value = 0

for i, val in enumerate(length_desc ):
    if val > max_value:
        max_value = val
        max_id = i
    if val < min_value:
        min_value = val
        min_id = i
    sum_value += val
```
